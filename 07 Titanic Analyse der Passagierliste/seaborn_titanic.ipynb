## 4. Datenbereinigung

```python
# Für weitere Analysen wird die Spalte "deck" nicht benötigt

df = df.drop(["deck"], axis = 1)
```

```python
# Weitere Spalten geben gleiche Werte wie "class" (kategorisch) und "pclass" (numerisch)

df = df.drop(["class"], axis = 1)
```

```python
# Desweiteren kann "adult_male" aus "who" abgeleitet werden

df = df.drop(["adult_male"], axis = 1)
```

```python
# Die Spalte "who" macht ebenfalls die Spalte "sex" redundant

df = df.drop(["sex"], axis = 1)
```

```python
# Die Spalten "alive" und "alone" wird nicht benötigt

df = df.drop(["alive"], axis = 1)
df = df.drop(["alone"], axis = 1)
```

```python
# Die Informationen aus "embark_town" sind ebenfalls in "embarked" enthalten

df = df.drop(["embark_town"], axis = 1)
```

```python
df.isnull().sum()
```

```python
# Alle 0 - Werte für das Alter werden durch den Durchschnittswert nach Geschlecht ("who" -> man, woman, child) ersetzt

# Durchschnittswert wird berechnet und alle 0 und NaN - Werte werden ausgeschlossen
average_age_man = df.loc[(df['who'] == 'man') & (df['age'] > 0), 'age'].mean()
average_age_woman = df.loc[(df['who'] == 'woman') & (df['age'] > 0), 'age'].mean()
average_age_child = df.loc[(df['who'] == 'child') & (df['age'] > 0), 'age'].mean()

# Durchschnittsalter von Männern, Frauen und Kindern
print(average_age_man)
print(average_age_woman)
print(average_age_child)

# Ersetzen von 0-Werten und NaN in 'age' durch berechnete Durchschnittswerte
df.loc[(df['who'] == 'man') & ((df['age'] == 0) | (df['age'].isna())), 'age'] = average_age_man
df.loc[(df['who'] == 'woman') & ((df['age'] == 0) | (df['age'].isna())), 'age'] = average_age_woman
df.loc[(df['who'] == 'child') & ((df['age'] == 0) | (df['age'].isna())), 'age'] = average_age_child
```

```python
df.isnull().sum()
```

```python
# Übrige 0 - Werte werden einfach gelöscht weil es nicht viele sind

df = df.dropna()
```

```python
df.isnull().sum()
```

## 6. Vorbereitung für Machine Learning

```python
df.sample(n = 15)
```

```python
# Aufteilen der Daten in "features" und "target"
features = ["pclass", "age","sibsp","parch", "fare","embarked","who"]
target = "survived"

# Umwandeln der kategorischen Werte zu numerischen Werten
data = df[features + [target]]
data['who'] = data['who'].replace(["woman", "man", "child"], [0, 1, 2])
data['embarked'] = data['embarked'].replace(['S', 'C', 'Q'], [1, 2, 3])
```

```python
data.sample(n = 15)
```

```python
corr = data.corr()
#cmap = sns.diverging_palette(230, 20, as_cmap=True)
plt.figure(figsize=(12, 8))
ax = sns.heatmap(corr,annot = True,fmt=".2f",linewidth=.2)
ax.set_title("Korrelationskoeffizienten")
plt.title('Korrelationsmatrix der Titanic-Daten')
plt.show()

#Ein Feature ist stark positiv korreliert, je näher die Zahl an 1 ist und stark negativ korreliert, je näher die Zahl an -1 ist.
```

## 7. Machine Learning

```python
# Bibliotheken für Machine Learning mit dem "Gaussian Naive Bayes"

from sklearn.metrics import accuracy_score, mean_absolute_error, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
# Aufteilen der Daten in Zielwerte y und Trainingswerte X

y = data["survived"]

X = data.drop(columns = ["survived"])

# Überprüfung der Werte

print(X.shape)
print(y.shape)
```

```python
# 1: Daten aufteilen in Trainings- und Testdaten
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# 2: Gaussian Naive Bayes Modell trainieren
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
# 3: Vorhersagen auf den Testdaten und Berechnung des Scores
y_pred = gnb.predict(X_test)
score = accuracy_score(y_test, y_pred)
```

```python
# Ausgabe des Scores
print(f"Genauigkeit des Naive Bayes Modells auf den Testdaten: {score:.2f}")
```
